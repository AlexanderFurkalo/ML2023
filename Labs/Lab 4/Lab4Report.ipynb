```python
import numpy as np 
import tensorflow as tf 

#1a
scalar_tensor = tf.constant(tf.random.uniform(shape=(), minval=0, maxval=1))
print("Scalar Tensor:")
print(scalar_tensor.numpy())

vector_tensor = tf.constant(tf.random.normal(shape=(5,)))
print("Vector Tensor:")
print(vector_tensor.numpy())

matrix_tensor = tf.constant(tf.random.normal(shape=(3, 4)))
print("Matrix Tensor:")
print(matrix_tensor.numpy())

variable_tensor = tf.Variable(tf.random.uniform(shape=(2, 3)))
print("Variable Tensor:")
print(variable_tensor.numpy())
#tf.constant - створює постійний тензор TensorFlow зі значенням, згенерованим випадковим рівномірним розподілом. 
#scalar tensor генерує випадкове скалярне значення (одне число), рівномірний розподілу між 0 і 1
#vector tensor генерує випадковий вектор (одновимірний масив) довжиною 5 із нормального розподілу
#  (розподіл Гауса із середнім значенням 0 і стандартним відхиленням 1).
#Variables можна змінювати під час тренування, constant є фіксованими.
#.numpy() — це метод, який дозволяє конвертувати тензор TensorFlow у масив NumPy.

#1b
tensdemo1 = tf.constant([1.0, 6.0, 4.0])
tensdemo2 = tf.constant([3.0, 5.0, 2.0])
tensdemo3 = tf.constant([0.0, 45.0, 90.0], dtype=tf.float32)
#dtype — це аргумент, який використовується для визначення типу даних елементів у постійному тензорі TensorFlow tensdemo3. 
# Встановлено на tf.float32, що означає, що елементи тензора мають тип 32-розрядних чисел з плаваючою комою.
addition_result = tensdemo1 + tensdemo2
print("\nAddition Result:")
print(addition_result.numpy())

subtraction_result = tensdemo1 - tensdemo2
print("Subtraction Result:")
print(subtraction_result.numpy())

multiplication_result = tensdemo1 * tensdemo2
print("Multiplication Result:")
print(multiplication_result.numpy())

division_result = tensdemo1 / tensdemo2
print("Division Result:")
print(division_result.numpy())

sine_result = tf.math.sin(tensdemo3)
print("Sine Result:")
print(sine_result.numpy())

cosine_result = tf.math.cos(tensdemo3)
print("Cosine Result:")
print(cosine_result.numpy())

#1c
element_1 = matrix_tensor[0, 0]
element_2 = matrix_tensor[1, 2]
print("\nElement 1: ", element_1.numpy())
print("Element 2: ", element_2.numpy())

row_1 = matrix_tensor[1, :]  
column_2 = matrix_tensor[:, 2]  
subset = matrix_tensor[0:2, 1:3]  
#Перше, 0:2 - це обрані стовпці, 1:3 - рядки
print("Row 1:")
print(row_1.numpy())
print("Column 2:")
print(column_2.numpy())
print("Subset:")
print(subset.numpy())

#1d
print("\nMatrix Tensor:")
print(matrix_tensor.numpy())

reshaped_tensor = tf.reshape(matrix_tensor, (1, 12))
# 1,12 - це форма, яку ми хочемо придати matrix_tensor за допомогою tf.reshape
print("Reshaped Tensor:")
print(reshaped_tensor.numpy())
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

num_samples = 1000
num_features = 2
features = np.random.rand(num_samples, num_features)
target_values = 2 * features[:, 0] + 3 * features[:, 1] + np.random.randn(num_samples)
# target_values -> там де будуть зберігатися згенеровані цільові дані.
# features[:, 0] - використовується для вилучення значень з першого стовпця feature
data = pd.DataFrame({'Feature_1': features[:, 0], 'Feature_2': features[:, 1], 'Target': target_values})
data.to_csv('Lab4Task2n3Data.csv') 

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(num_features,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  
])
# tf.keras.Sequential створює порожню послідовну модель. 
#Послідовна модель — це лінійний стек шарів, і це загальний вибір для побудови прямої нейронної мережі.
# tf.keras.layers.Dense(64) - додає повністю пов’язаний (щільний) шар до моделі з 64 нейронами або одиницями. 
#Кожен нейрон цього шару з’єднаний з усіма нейронами попереднього шару.
# relu = Rectified Linear Unit, функція активації
# input_shape - визначає вхідну форму для першого шару. 
#Визначає вхідний рівень із тією ж кількістю нейронів, що й кількість функцій у ваших даних.
# tf.keras.layers.Dense(1) - останній рівень виводу з одним нейроном, застосовується для завдань з регресією

model.compile(optimizer='adam', loss='mean_squared_error')
# optimizer - визначає алгоритм оптимізації, який буде використовуватися під час навчання. 
#ada поєднує ідеї з двох інших алгоритмів оптимізації: AdaGrad і RMSprop, 
#і він відомий своєю ефективною конвергенцією та високою продуктивністю для різних завдань.
# loss - loss, функція втрат, яку модель оптимізує під час навчання
features_train, features_val, target_train, target_val = train_test_split(features, target_values, test_size=0.2, random_state=42)
history = model.fit(features_train, target_train, epochs=100, validation_data=(features_val, target_val))
# epochs - визначає кількість ітерацій або епох навчання. У цьому випадку модель проходитиме тренувальні дані 100 разів,
#регулюючи свої параметри (ваги та зміщення) для покращення продуктивності.
# history - зберігає історію навчання моделі

target_pred = model.predict(features_val)
mae = tf.keras.losses.MeanAbsoluteError()(target_val, target_pred).numpy()
print("Mean Absolute Error:", mae)
mse = tf.keras.losses.MeanSquaredError()(target_val, target_pred).numpy()
print("Mean Squared Error (MSE):", mse)

print("\nData visualization")
plt.scatter(data['Feature_1'], data['Feature_2'], c=data['Target'], cmap='viridis', alpha=0.5)
# c=data['Target'] - визначає колір кожної точки даних на діаграмі розсіювання на основі змінної 
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Features 1 and 2 (Color Encodes Target)')
plt.colorbar()
plt.show()

plt.hist(data['Feature_1'], bins=10, alpha=0.5, color='green', label="Histogram")
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.title('Histogram of Feature 1')
plt.legend()
plt.show()

feature_columns = [col for col in data.columns if col != 'Target']
for feature in feature_columns:
    plt.plot(data[feature], label=feature)
plt.xlabel('Data Points')
plt.ylabel('Feature Values')
plt.title('Line Graph of All Features')
plt.legend()
plt.show()

print("\nResults visualization")
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(target_val, target_pred, alpha=0.5)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.text(0.1, 0.9, f'MAE: {mae:.2f}', transform=plt.gca().transAxes, fontsize=12)
plt.text(0.1, 0.85, f'MSE: {mse:.2f}', transform=plt.gca().transAxes, fontsize=12)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split

num_samples = 1000
num_features = 2
features = np.random.rand(num_samples, num_features)
labels = np.random.randint(2, size=num_samples)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(num_features,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid') 
])
# Для останнього рядка вказано activation - вихідні дані цього шару передаються через функцію активації sigmoid.
#зазвичай використовується для двійкової класифікації, оскільки вона здавлює вихідні дані в діапазон [0, 1],
#що представляє ймовірність приналежності до позитивного класу.

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
features_train, features_val, labels_train, labels_val = train_test_split(features, labels, test_size=0.2, random_state=42)
history = model.fit(features_train, labels_train, epochs=20, validation_data=(features_val, labels_val))

test_loss, test_accuracy = model.evaluate(features_val, labels_val)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```
