Завдання 3 (Matplotlib)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
np.random.seed(0)

slope = 2.0
#Нахил slope представляє крутизну або нахил лінії
intercept = 5.0
#Точка перетину є точкою, в якій лінія перетинає вісь y
x = np.linspace(0, 10, 20) 
#Використовується для створення масиву рівномірно розподілених значень за вказаний інтервал - 20 значень, 0-10
y_true = slope * x + intercept  
#y = mx + b, linear equation
measurement_errors = np.random.normal(0, 2, len(x))  
#Генерує серію випадкових чисел (помилок вимірювання),
#які відповідають нормальному розподілу із середнім значенням 0 і стандартним відхиленням 2.
#len(x) = кількість випадкових значень для генерації, ми створимо стільки "помилок", скільки маємо точок даних у масиві 
y = y_true + measurement_errors
plt.scatter(x, y, label='Data with Errors', color='blue', marker='o')
plt.plot(x, y_true, label='True Linear Function', color='red', linestyle='--')
#Використовується для візуалізації даних шляхом нанесення точок і з’єднання їх лініями
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Function with Measurement Errors')
plt.legend()
plt.grid(True)
mae = mean_absolute_error(y_true, y)
#"MAE" означає середню абсолютну похибку
mse = mean_squared_error(y_true, y)
#"MSE" означає середню квадратичну помилку. Підкреслює більші помилки більше, ніж менші завдяки операції зведення в квадрат
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
plt.show()
results = pd.DataFrame({'X': x, 'Y': y, 'Y_hat': y_true, 'MAE': mae, 'MSE': mse})
print(results)
results.to_csv('results.csv', index=False)
```
