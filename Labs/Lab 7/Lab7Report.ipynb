```python
import os
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import tensorflow_hub as hub
from keras.callbacks import EarlyStopping
import time

classes = ['chicken_wings', 'ice_cream', 'sushi']

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

test_datagen = ImageDataGenerator(rescale=1./255)

project_path = '../../'
data_path = '../../../ML2023/Practices/Practice8/10_food_classes_all_data'
print("Project Path:", project_path)
print("Data Path:", data_path)
train_data_path = os.path.join(data_path, 'train')
test_data_path = os.path.join(data_path, 'test')

train_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),  
    batch_size=32,
    classes=classes,
    class_mode='categorical',  
    subset='training'  
)

validation_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    classes=classes,
    class_mode='categorical',
    subset='validation' 
)

test_generator = test_datagen.flow_from_directory(
    test_data_path,
    target_size=(224, 224),
    batch_size=32,
    classes=classes,
    class_mode='categorical'
)

model1 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), input_shape=(224, 224, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax') 
])
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model2 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), input_shape=(224, 224, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model3 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), input_shape=(224, 224, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

efficientnet_url = "https://tfhub.dev/tensorflow/efficientnet/b0/feature-vector/1"
efficientnet_model = tf.keras.Sequential([
    hub.KerasLayer(efficientnet_url, trainable=False),
    tf.keras.layers.Dense(3, activation='softmax')
])
efficientnet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time = time.time()
history1 = model1.fit(train_generator, validation_data=validation_generator, epochs=10, callbacks=[EarlyStopping(patience=3)])
training_time1 = time.time() - start_time

start_time = time.time()
history2 = model2.fit(train_generator, validation_data=validation_generator, epochs=10, callbacks=[EarlyStopping(patience=3)])
training_time2 = time.time() - start_time

start_time = time.time()
history3 = model3.fit(train_generator, validation_data=validation_generator, epochs=10, callbacks=[EarlyStopping(patience=3)])
training_time3 = time.time() - start_time

eval1 = model1.evaluate(test_generator)
eval2 = model2.evaluate(test_generator)
eval3 = model3.evaluate(test_generator)

print("Model 1 - Accuracy:", eval1[1], "Training Time:", training_time1)
print("Model 2 - Accuracy:", eval2[1], "Training Time:", training_time2)
print("Model 3 - Accuracy:", eval3[1], "Training Time:", training_time3)

start_time = time.time()
history_efficientnet = efficientnet_model.fit(train_generator, validation_data=validation_generator, epochs=10, callbacks=[EarlyStopping(patience=3)])
training_time_efficientnet = time.time() - start_time

eval_efficientnet = efficientnet_model.evaluate(test_generator)

print("EfficientNet - Accuracy:", eval_efficientnet[1], "Training Time:", training_time_efficientnet)
```
